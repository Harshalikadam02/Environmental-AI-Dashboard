### Combined Datasets

```python
df_traffic_counts['Date'] = pd.to_datetime(df_traffic_counts['Date']).dt.date
df_traffic_counts.head()
```

```python
import pandas as pd
df_combined = pd.merge(
    df_air_quality,
    df_traffic_counts,
    left_on=['Start_Date', 'Boro'],
    right_on=['Date', 'Boro'],
    how='inner'
)
df_combined.head()
```

```python
columns_to_drop_combined = ['Yr', 'M', 'D', 'HH', 'MM', 'DateTime', 'Date']
df_combined = df_combined.drop(columns=columns_to_drop_combined)
```

```python
df_combined.head()
```

###Analysis

```python
df_combined['Name'].unique()
```

```python
df_pollution_trends = df_combined.groupby(['Boro', 'Start_Date', 'Name']).agg(
    mean_data_value=('Data Value', 'mean'),
    mean_vol=('Vol', 'mean')
).reset_index()
df_pollution_trends.head()
```

```python
correlation_by_pollutant = df_pollution_trends.groupby('Name').apply(lambda x: x['mean_data_value'].corr(x['mean_vol'], method='pearson'))
print("Pearson correlation between pollution levels and traffic volume by pollutant type:\n")
print(correlation_by_pollutant)
```

```python
air_quality_thresholds = {
    'Fine particles (PM 2.5)': 12,
    'Nitrogen dioxide (NO2)': 50,
    'Ozone (O3)': 70

df_poor_air_quality_list = []

for pollutant, threshold in air_quality_thresholds.items():

    df_exceeding_threshold = df_pollution_trends[
        (df_pollution_trends['Name'] == pollutant) &
        (df_pollution_trends['mean_data_value'] > threshold)
    ]
    df_poor_air_quality_list.append(df_exceeding_threshold)

df_poor_air_quality = pd.concat(df_poor_air_quality_list)

print("Periods of Poor Air Quality (first 5 rows):")
print(df_poor_air_quality.head())
print(f"\nTotal identified periods of poor air quality: {len(df_poor_air_quality)}")
```
